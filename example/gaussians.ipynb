```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

from sklearn.mixture import GaussianMixture
```

```python
class_size = 500
class_count = 5

samples = np.vstack([
    np.random.normal(0, 0.3, size=(class_size, 2)) + np.array([1, 1]),
    np.random.normal(0, 0.5, size=(class_size, 2)) + np.array([2, 1]),
    np.random.normal(0, 0.7, size=(class_size, 2)) + np.array([4, 3]),
    np.random.normal(0, 0.9, size=(class_size, 2)) + np.array([3, 1]),
    np.random.normal(0, 0.6, size=(class_size, 2)) + np.array([0, 2]),
])
```

```python
gmm = GaussianMixture(n_components=class_count)
gmm.fit(samples)
probs = gmm.predict_proba(samples)
```

```python
label_colors = np.array([
    (1.0, 0.2, 0.3),
    (0.3, 1.0, 0.0),
    (0.3, 0.6, 1.0),
    (1.0, 0.9, 0.0),
    (0.8, 0.1, 1.0),
])
colors = probs.dot(label_colors)
```

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi=96)

ax.scatter(samples[:, 0], samples[:, 1], lw=0, s=15, c=colors, alpha=0.9, zorder=2)

ax.patch.set_facecolor("#333333")
ax.set_aspect("equal")
ax.grid(True, lw=1, ls=":", color="#cccccc", alpha=0.5, zorder=1)

pass
```
